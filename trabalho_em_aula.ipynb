```python
import pandas as Po
notas = Po.Series([7.5, 8.0, 9.0, 6.5, 8.5])
#========================================================
print("retorna o numero de indices do array, ou seja, a quantidade de elementos:")
print(notas.index)
#========================================================
print("retorna os valores do array, ou seja, as notas:")
print(notas.values)
#========================================================
print("retorna os 3 primeiros elementos do array, ou seja, as notas 7.5, 8.0 e 9.0:")
print(notas.head(n=3))
#========================================================
print("retorna os 2 ultimos elementos do array, ou seja, as notas 8.5 e 6.5:")
print(notas.tail(2))
print("retorna a quantidade de elementos do array, ou seja, 5")
print(notas.shape)#retorna a quantidade de elementos do array, ou seja, 5
```
